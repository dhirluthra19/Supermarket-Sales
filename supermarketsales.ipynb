## DATA CLEANING

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')
df['Date'] = pd.to_datetime(df['Date'] , format ='mixed')
```

```python
#Checking Duplicates
df.duplicated(subset=["Branch","City","Customer type","Date","Time"]).sum()
```

```python
df.duplicated().sum()
```

## ADDING EXTRA COLUMN FOR  ANALYSIS 

```python
df["Hour"]=df["Time"].dt.hour
df["Month"]=df["Date"].dt.month_name()
```

```python
df['Branch'] = df[['Branch', 'City']].agg(' Branch in city '.join, axis=1)
```

## UNIQUE VALUE IN ALL CATEGORICAL COLUMN


```python
cols = [col for col in df.columns if df[col].dtype =="O" and col !="Invoice ID"]
       
```

```python
for col in cols:
    print(f"{cols}:{df[col].unique()}    number:{df[col].nunique()}")
```

## GRAPHICAL CATEGORICALS COLUMNS

```python
plt.figure(figsize=(28,23))
for i, col in enumerate(cols):
    plt.subplot(5, 2, i+1)
    sns.histplot(data = df,
            x = col,
            kde = True,
            bins = 30,
            color = 'red')

plt.show()
```
